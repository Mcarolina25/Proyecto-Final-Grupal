```python
import os
import json
import pandas as pd

def cargar_reviews_desde_directorio(ruta_directorio):
    """Carga reviews desde todos los archivos JSON en un directorio a una lista de diccionarios."""
    reviews = []
    for filename in os.listdir(ruta_directorio):
        if filename.endswith(".json"):
            ruta_archivo = os.path.join(ruta_directorio, filename)
            try:
                with open(ruta_archivo, 'r', encoding='utf-8') as f:
                    for line in f:
                        try:
                            review = json.loads(line)
                            reviews.append(review)
                        except json.JSONDecodeError:
                            print(f"Omitiendo línea JSON inválida en {filename}: {line.strip()}")
            except FileNotFoundError:
                print(f"Error: Archivo no encontrado: {ruta_archivo}")
            except json.JSONDecodeError:
                print(f"Error decodificando JSON en el archivo: {ruta_archivo}")
            except Exception as e:
                print(f"Ocurrió un error inesperado al procesar {ruta_archivo}: {e}")
    return reviews

# Ejemplo de uso:
ruta_directorio = r'c:\Users\segen\Documents\Henry\Proyecto3-Grupo\reviews-estados\review-Michigan'  # Reemplaza con la ruta real a tu directorio
reviews = cargar_reviews_desde_directorio(ruta_directorio)

# Crea un DataFrame de Pandas desde la lista de reviews
df = pd.DataFrame(reviews)

# Imprime el DataFrame para verificar los datos
print(df)
```

```python
def cargar_reviews_desde_directorio(ruta_directorio):
    """Carga reseñas desde todos los archivos JSON en un directorio a un DataFrame."""
    reviews = []
    for filename in os.listdir(ruta_directorio):
        if filename.endswith(".json"):
            ruta_archivo = os.path.join(ruta_directorio, filename)
            try:
                with open(ruta_archivo, 'r', encoding='utf-8') as f:
                    for line in f:
                        try:
                            review = json.loads(line)
                            reviews.append(review)
                        except json.JSONDecodeError:
                            print(f"Omitiendo línea JSON inválida en {filename}: {line.strip()}")
            except FileNotFoundError:
                print(f"Error: Archivo no encontrado: {ruta_archivo}")
            except json.JSONDecodeError:
                print(f"Error decodificando JSON en el archivo: {ruta_archivo}")
            except Exception as e:
                print(f"Ocurrió un error inesperado al procesar {ruta_archivo}: {e}")
    return pd.DataFrame(reviews)

# Lista de directorios de estados
estados = ['Michigan', 'Kansas', 'Kentucky', 'Mississippi', 'Missouri', 'Montana', 'Nebraska', 'Nevada', 'New_Hampshire', 'New_York', 'North_Carolina', 'Ohio', 'Oklahoma', 'Oregon', 'Pennsylvania', 'Rhode_Island', 'South_Dakota', 'Tennessee', 'Utah', 'Vermont', 'Wisconsin']

# Ruta base del directorio
ruta_base = r'c:\Users\segen\Documents\Henry\Proyecto3-Grupo\reviews-estados'

# Diccionario para almacenar los DataFrames
dataframes = {}

# Cargar los datos de cada estado
for estado in estados:
    ruta_directorio = os.path.join(ruta_base, f'review-{estado}')
    try:
        dataframes[estado] = cargar_reviews_desde_directorio(ruta_directorio)
        print(f"DataFrame para {estado} cargado con éxito.")
    except Exception as e:
        print(f"Error al cargar DataFrame para {estado}: {e}")

# Ahora puedes acceder a los DataFrames individuales usando el diccionario 'dataframes'
# Por ejemplo: dataframes['Michigan'] te dará el DataFrame de Michigan
```
